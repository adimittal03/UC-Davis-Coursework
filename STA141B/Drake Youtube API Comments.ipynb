```python
from googleapiclient.discovery import build
from collections import defaultdict
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
api_key = "AIzaSyCKhyPTPEHgjrobMjxXQOAq-Q-X5YyKRfc"
video_id = "xpVfcZ0ZcFM"  #extracting from God's Plan
#build a resource for youtube ??
resource = build('youtube', 'v3', developerKey=api_key)

#create a request to get 100 comments on the video
request = resource. commentThreads().list(
                            part="snippet",
                            videoId=video_id,
                            maxResults= 100,
                            order="orderUnspecified") #top comments

#execute the request
response = request.execute()
```

```python
response.keys()
```

```python
items = response['items'][:99]
items
df=pd.json_normalize(items)
df=df['snippet.topLevelComment.snippet.textDisplay']
df.head()
```

```python
text = df.to_csv() #converting the data frame to text file 
# in order to generate to a word cloud of a shape in python
```

```python
import nltk
from nltk import tokenize
words = tokenize.word_tokenize(text)
words = [w for w in words if w.isalpha()] #assures only words are in the output 
#words=[x.lower() for x in words] #lower-casing the words as well 
```

```python
#Removing stopwords 
stopwords = ["the", "a", "and", "or", "in"]
final_words = [w for w in words if w not in stopwords]
```

```python
type(final_words)
```

```python
text = ' '.join(final_words) #turning list back into string
type(text)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
from PIL import Image
```

```python
stopwords = set(STOPWORDS)
cloud = np.array(Image.open("cloud.png"))
# Create a word cloud image
wc = WordCloud(background_color="black", max_words=1000, mask=cloud,
               contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(text)

# store to file

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
